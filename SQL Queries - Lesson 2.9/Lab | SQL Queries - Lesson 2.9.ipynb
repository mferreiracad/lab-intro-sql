```python
# How many distinct (different) actors' last names are there?
```

```python
query = '''select count(distinct a.last_name) as "Number of distinct last names"
from actor a;'''

n_actors = pd.read_sql_query(query, engine)
```

```python
#Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
```

```python
query = """SELECT *,
date_format(CONVERT(LEFT(rental_date,23),DATE), '%%M') AS 'Month',
date_format(CONVERT(LEFT(rental_date,23),DATE), '%%W') AS 'Weekday',
CASE
WHEN date_format(CONVERT(left(rental_date,23),DATE), '%%W') = ('Tuesday' | 'Monday' | 'Wednesday' | 'Friday' | 'Thursday') then 'Workday'
ELSE 'Weekend'
END AS 'day_type'
FROM sakila.rental 
;"""

rental = pd.read_sql_query(query, engine)
rental
```

```python
# Get all films with ARMAGEDDON in the title.
```

```python
query = """SELECT * FROM sakila.film WHERE title LIKE '%%ARMAGEDDON%%'"""

armageddon = pd.read_sql_query(query, engine)
armageddon
```

```python
# Get 10 the longest films.

query = """SELECT film.title, film.length FROM sakila.film ORDER BY length DESC LIMIT 10;"""

longest = pd.read_sql_query(query, engine)
longest
```

```python
# How many films include Behind the Scenes content?
query = """SELECT COUNT(*) FROM sakila.film WHERE special_features LIKE '%%Behind the Scenes%%';"""

behind_scenes_count = pd.read_sql_query(query, engine)
behind_scenes_count
```

```python
# Which kind of movies (rating) have a mean duration of more than two hours?

query = """SELECT sakila.film.rating, AVG(sakila.film.length) AS Average
FROM sakila.film
GROUP BY sakila.film.rating
HAVING Average > 120;"""

more_t_2 = pd.read_sql_query(query, engine)
more_t_2
```

```python
#Rank films by length (filter out the rows that have nulls or 0s in length column).
# In your output, only select the columns title, length, and the rank.
query = """SELECT sakila.film.title, sakila.film.length, DENSE_RANK() OVER(ORDER BY length DESC) AS 'Rank'
FROM sakila.film
WHERE length IS NOT NULL AND length != 0
ORDER BY length DESC;"""

ranking = pd.read_sql_query(query, engine)
ranking
```
